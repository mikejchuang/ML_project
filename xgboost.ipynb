```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns', 100)
```

```python
test_dummy = pd.read_csv('./Data/cleaned_test_dummy.csv')
train_dummy = pd.read_csv('./Data/cleaned_train_dummy.csv')
```

```python
train_dummy.head(30)
```

```python
from sklearn.metrics import mean_squared_error
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))
```

```python
import xgboost as xgb
```

```python
prep_X = train_dummy.drop(['Unnamed: 0', 'Id', 'SalePrice'], axis =1)
prep_y = train_dummy.SalePrice
```

```python
prep_test = test_dummy.drop(['Unnamed: 0', 'Id', 'SalePrice'], axis =1)
```

```python
regr = xgb.XGBRegressor(
                 colsample_bytree=0.2,
                 gamma=0.0,
                 learning_rate=0.001,
                 max_depth=5,
                 min_child_weight=1.5,
                 n_estimators=1000,                                                                  
                 reg_alpha=0.9,
                 reg_lambda=0.6,
                 subsample=0.2,
                 seed=1,
                 silent=False)

regr.fit(prep_X, prep_y)

# Run prediction on training set to get a rough idea of how well it does.
y_pred = regr.predict(prep_X)

print("RMSE on training set: ", rmse(prep_y, y_pred))
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
regr = xgb.XGBRegressor()
```

```python
parameters = {
    'max_depth': [3,4,5,6,7],
    'learning_rate': [.0001, .001, .01, .1],
    'n_estimators': [200],
    'gamma': [0.1],
    'reg_alpha': [.25, .5, .75],
    'reg_lambda': [.25, .5, .75],
    'subsample': [.3 ,.5,.7,.9],
    'colsample_bytree': [.4,.5,.6],
    'min_child_weight': [.3, .6, .9]
    }
```

```python
tuning = GridSearchCV(regr, parameters,scoring='neg_mean_squared_error', cv=5, n_jobs=-1, verbose=1)
tuning.fit(prep_X, prep_y)
```

```python
tuning.best_score_
```

```python
new_regr = xgb.XGBRegressor(
                 nthread=4,
                 colsample_bytree=0.6,
                 gamma=0.0,
                 learning_rate=0.001,
                 max_depth=4,
                 min_child_weight=0.3,
                 n_estimators=100000,                                                                  
                 reg_alpha=0.25,
                 reg_lambda=0.25,
                 subsample=0.3,
                 seed=1,
                 silent=False)

new_regr.fit(prep_X, prep_y)
```

```python
y_pred = new_regr.predict(prep_X)

print("RMSE on training set: ", rmse(prep_y, y_pred))
```
